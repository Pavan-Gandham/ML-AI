```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import  Dense
```

```python
data = pd.read_csv("/content/USA_Housing_syn.csv")
```

```python
X= df.values[:,0:3]
Y=  df.values[:,4]
Y=np.reshape(Y, (-1,1))
print(X.shape)
print(Y.shape)
```

```python
model=Sequential()
model.add(Dense(12,input_dim=3,activation='relu')) #check the input dimension(number of inputs) is important
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='linear'))
model.summary()
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['mse','mae'])
```

```python
x_train, x_test,y_train, y_test = train_test_split(X,Y,random_state=104,test_size=0.25)
```

```python
#oob_score = True, n_jobs = 1,random_state =50,max_features = "auto", min_samples_leaf =50
ranfor = RandomForestRegressor()
# ttr = TransformedTargetRegressor(regressor=ranfor, transformer=QuantileTransformer(output_distribution='normal'))
ranfor.fit(x_train,y_train.ravel()) #ravel will flatten the array
#ranfor_predict=ranfor.predict(x_test)
#ranfor_predict.reshape(-1)
#ranfor_predict.shape
```

```python
history= model.fit(x_train,y_train,epochs=150,batch_size=20)
```

```python
xnew=np.array([[80000,8,6]])
ynew= ranfor.predict(xnew)
print("X=%s, Predicted=%s" % (xnew[0], ynew[0]))
```

```python
import pickle
with open('housing.pkl', 'wb') as f:
    pickle.dump(ranfor, f)
```
